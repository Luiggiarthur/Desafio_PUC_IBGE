```python
df_cidades.columns
```

```python
df_cidades_sum = df_cidades[['cod_uf', 'qtd_dom', 'qtd_dom_pobres',
       'qtd_dom_vulneraveis', 'qtd_dom_pob_vul', 'qtd_pes', 'qtd_pes_pobres',
       'qtd_pes_vulneraveis', 'qtd_pes_pob_vul', 'qtd_0a11', 'qtd_0a11_pobres',
       'qtd_0a11_vulneraveis', 'qtd_0a11_pob_vul','qtd_dom_E1', 'qtd_dom_E2',
       'qtd_dom_E3', 'qtd_dom_S1', 'qtd_dom_S2', 'qtd_dom_S3', 'qtd_dom_S4',
       'qtd_dom_T1', 'qtd_dom_T2', 'qtd_dom_T3', 'qtd_dom_P1', 'qtd_dom_P2',
       'qtd_dom_P3', 'qtd_dom_P4',]]

df_cidades_mean = df_cidades[['cod_uf','incidencia', 'intensidade','IPM', 'num_idhm','num_ivs', 'num_ibeu', 'num_renda']]
```

```python
df_estados_1 = df_cidades_sum.groupby('cod_uf').sum().reset_index()
df_estados_1[['qtd_dom', 'qtd_dom_pobres',
       'qtd_dom_vulneraveis', 'qtd_dom_pob_vul', 'qtd_pes', 'qtd_pes_pobres',
       'qtd_pes_vulneraveis', 'qtd_pes_pob_vul', 'qtd_0a11', 'qtd_0a11_pobres',
       'qtd_0a11_vulneraveis', 'qtd_0a11_pob_vul','qtd_dom_E1', 'qtd_dom_E2',
       'qtd_dom_E3', 'qtd_dom_S1', 'qtd_dom_S2', 'qtd_dom_S3', 'qtd_dom_S4',
       'qtd_dom_T1', 'qtd_dom_T2', 'qtd_dom_T3', 'qtd_dom_P1', 'qtd_dom_P2',
       'qtd_dom_P3', 'qtd_dom_P4',]] = df_estados_1[['qtd_dom', 'qtd_dom_pobres',
       'qtd_dom_vulneraveis', 'qtd_dom_pob_vul', 'qtd_pes', 'qtd_pes_pobres',
       'qtd_pes_vulneraveis', 'qtd_pes_pob_vul', 'qtd_0a11', 'qtd_0a11_pobres',
       'qtd_0a11_vulneraveis', 'qtd_0a11_pob_vul','qtd_dom_E1', 'qtd_dom_E2',
       'qtd_dom_E3', 'qtd_dom_S1', 'qtd_dom_S2', 'qtd_dom_S3', 'qtd_dom_S4',
       'qtd_dom_T1', 'qtd_dom_T2', 'qtd_dom_T3', 'qtd_dom_P1', 'qtd_dom_P2',
       'qtd_dom_P3', 'qtd_dom_P4',]].div(df_estados_1['qtd_pes'], axis='index')
df_estados_1
```

```python
df_estados_2 = df_cidades_mean.groupby('cod_uf').mean().reset_index()
df_estados_2
```

```python
df_merged = df_estados_2.merge(df_estados_1)
```

```python
df_IPM_cidade=df_merged[['cod_uf','IPM']].sort_values(by='IPM')
mapping = {
    11: 'RO',
    12: 'AC',
    13: 'AM',
    14: 'RR',
    15: 'PA',
    16: 'AP',
    17: 'TO',
    21: 'MA',
    22: 'PI',
    23: 'CE',
    24: 'RN',
    25: 'PB',
    26: 'PE',
    27: 'AL',
    28: 'SE',
    29: 'BA',
    31: 'MG',
    32: 'ES',
    33: 'RJ',
    35: 'SP',
    41: 'PR',
    42: 'SC',
    43: 'RS',
    50: 'MS',
    51: 'MT',
    52: 'GO',
    53: 'DF'
}

df_IPM_cidade['UF'] =df_IPM_cidade['cod_uf'].map(mapping)
df_IPM_cidade.to_excel('IPM_cidade.xlsx')
df_IPM_cidade
```

```python

from sklearn.preprocessing import MinMaxScaler
df_scaled = df_merged.drop('cod_uf',axis = 1)
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_scaled)
df_scaled =pd.DataFrame(df_scaled,columns=df_merged.drop('cod_uf',axis = 1).columns)
df_scaled = pd.concat([df_scaled,df_merged['cod_uf']], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(df_merged[['cod_uf']])
encoded_df = pd.DataFrame(encoded_data)
```

```python
df_merged = pd.concat([df_scaled,encoded_df], axis=1)
df_merged
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,28):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(df_merged)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
ssd    
```

```python
plt.plot(range(2,28),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
model = KMeans(n_clusters=5)
model.fit(df_merged)
```

```python
model.labels_
```

```python
df_merged['K=5 Clusters'] = model.labels_
```

```python
df_cluster = df_merged[['K=5 Clusters','cod_uf']]

df_cluster['UF'] =df_cluster['cod_uf'].map(mapping)
df_cluster.to_excel('cluster.xlsx')
df_cluster
```
